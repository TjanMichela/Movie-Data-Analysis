__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).

* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.

* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

```python
# Create a copy of relevant data into a new DataFrame movies_data_best
movies_data_best = movies_data[["poster_path", "title", "budget_musd", "revenue_musd", "vote_count", "vote_average", "popularity"]].copy()

movies_data_best
```

```python
# Create "profit_musd" Series revenue_musd minus budget_musd
movies_data_best["profit_musd"] = movies_data["revenue_musd"]-(movies_data.budget_musd)

# Create "return" Series revenue_musd divided by budget_musd
movies_data_best["return"] = movies_data["revenue_musd"]/(movies_data.budget_musd)
```

__Define__ an appropriate __user-defined function__ to reuse code.

```python
def best_worse(n, by, ascending=False, min_budget=0, min_votes=0):
    
    movies_data_2 = movies_data_best.loc[(movies_data_best["Budget"] >= min_budget) & (movies_data_best["Votes"] >= min_votes), ["", by]].sort_values(by=by, ascending=ascending).head(n).copy()

    return HTML(movies_data_2.to_html(escape=False))
```
